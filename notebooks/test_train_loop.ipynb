```python
import os
from pathlib import Path
import csv

# Set your target directory here
root_dir = "/mnt/data/StressID-img-mtcnn-data/"
output_csv = Path(root_dir) / "dataset2.csv"


with open(output_csv, 'w', newline='') as f:
    writer = csv.writer(f, delimiter=';')
    writer.writerow(["path", "str-label", "label"])

    for png_file in Path(root_dir).rglob("*.png"):
        path = str(png_file.resolve()).lower()

        if "no-stress" in path:
            str_label = "no-stress"
            label = 0
        elif "stress" in path:
            str_label = "stress"
            label = 1
        # else:
        #     str_label = "unknown"
        #     label = -1  # or skip the file

        writer.writerow([str(png_file.resolve()), str_label, label])

print(f"CSV saved to {output_csv}")   
```

```python
from sklearn.model_selection import train_test_split
import pandas as pd

# Set your target directory here
root_dir = "/mnt/data/StressID-img-mtcnn-data/"
output_csv = Path(root_dir) / "dataset2.csv"
output_csv_train = Path(root_dir) / "train.csv"
output_csv_test = Path(root_dir) / "test.csv"


# Load the CSV
df = pd.read_csv(output_csv, delimiter=';')

# Split the full DataFrame
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df['label'])

# Save to CSV
train_df.to_csv(output_csv_train, index=False, sep=';')
test_df.to_csv(output_csv_test, index=False, sep=';')

print("Train and test splits saved.")   
```
