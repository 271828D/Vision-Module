```python
from pathlib import Path
import pandas as pd

path = "/mnt/data/StressID-img-mtcnn-data/"
folder_path = Path(path)

with_face = []
without_face = []

for file in folder_path.rglob('*.png'):
    str_label = file.parent.parent.parent.name
    if "face" in file.name:
        with_face.append({'path': str(file), 'str-label': str_label})
    else:
        without_face.append({'path': str(file), 'str-label': str_label})

# Create DataFrames
df_with_face = pd.DataFrame(with_face)
df_without_face = pd.DataFrame(without_face)

# Optional: Map string labels to numeric
label_map = {'stress': 1.0, 'no-stress': 0.0}
df_with_face['label'] = df_with_face['str-label'].map(label_map)
df_without_face['label'] = df_without_face['str-label'].map(label_map)

# Display results
print("With 'face' in filename:")
print(df_with_face.shape)
print("\nWithout 'face' in filename:")
print(df_without_face.shape)   
```

```python
from pathlib import Path
import shutil

src_path = "/mnt/data/StressID-img-mtcnn-data/"
dst_path = "/mnt/data/StressID-img-mtcnn2-data/"  # Choose your destination

Path(dst_path).mkdir(exist_ok=True)

for file in Path(src_path).rglob('*.png'):
    if "face" not in file.name:
        dest_file = Path(dst_path) / file.relative_to(src_path)
        dest_file.parent.mkdir(exist_ok=True, parents=True)
        shutil.move(str(file), str(dest_file))   
```

```python
from sklearn.model_selection import train_test_split

# Assuming your DataFrame is called 'df'
train_df, test_df = train_test_split(df_without_face, test_size=0.1, random_state=42, shuffle=True)



# Save to CSV
df_without_face.to_csv(os.path.join(path, 'dataset.csv'), sep=";", index=False)
train_df.to_csv(os.path.join(path, 'train.csv'), sep=";", index=False)
test_df.to_csv(os.path.join(path, 'test.csv'), sep=";", index=False)

print(f"\r df shape: {df_without_face.shape}")
print(f"\r train shape: {train_df.shape}")
print(f"\r test shape: {test_df.shape}")
```
